```python
from sqlalchemy import create_engine
import pandas as pd
import datetime
from datetime import date
from datetime import timedelta

link_glo_con = r"C:\Users\Victor\DAG\covidtest\COVID-19\csse_covid_19_data\csse_covid_19_time_series\time_series_covid19_confirmed_global.csv"
link_glo_death = r"C:\Users\Victor\DAG\covidtest\COVID-19\csse_covid_19_data\csse_covid_19_time_series\time_series_covid19_deaths_global.csv"
link_glo_rec = r"C:\Users\Victor\DAG\covidtest\COVID-19\csse_covid_19_data\csse_covid_19_time_series\time_series_covid19_recovered_global.csv"
dfconf = pd.read_csv(link_glo_con)
dfded = pd.read_csv(link_glo_death)
dfrec = pd.read_csv(link_glo_rec)
```

```python
dfconf.drop(['Lat', 'Long'], axis=1, inplace=True)
dfded.drop(['Lat', 'Long'], axis=1, inplace=True)
dfrec.drop(['Lat', 'Long'], axis=1, inplace=True)
dfconf
```

```python
if not date_ranges:
    print('No new data')
else:
    date_ranges.pop(0)
    if len(date_ranges)==0:
        date_ranges = x.strftime("%x")
        date_ranges = date_ranges.replace('/', '-')
    else:
        date_ranges=[i.replace('/', '-') for i in date_ranges]
date_ranges
```

```python
def format_date(date_str):
    new_date = str()
    if int(date_str[:2]) < 10:
        new_date += date_str[:2].replace('0','')
        new_date+='/'
    else:
        new_date+=date_str[:2]
        new_date+='/'

    if int(date_str[3:5]) < 10:
        new_date += date_str[3:5].replace('0','')
        new_date+='/'
    else:
        new_date+=date_str[3:5]
        new_date+='/'

    new_date+=date_str[-2:]
    return new_date
```
